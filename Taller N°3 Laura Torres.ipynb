## Punto No 1

```python
##Importar Librerias
import urllib.request
import bs4 as bs
```

```python
#Definir los links
animal1 ="https://es.wikipedia.org/wiki/Crocodylidae"
animal2 ="https://es.wikipedia.org/wiki/Eunectes_murinus"
animal3 ="https://es.wikipedia.org/wiki/Anisoptera"
animal4 ="https://es.wikipedia.org/wiki/Mosquito"
animal5 ="https://es.wikipedia.org/wiki/Elephantidae"
animal6 ="https://es.wikipedia.org/wiki/Hyaenidae"
animal7 ="https://es.wikipedia.org/wiki/Lycaon_pictus"
animal8 ="https://es.wikipedia.org/wiki/Hippocampus"
animal9 ="https://es.wikipedia.org/wiki/Mobula_birostris"
animal10 ="https://es.wikipedia.org/wiki/Fulgora_laternaria"
```

```python
#Hacer una lista con los links
listado =[animal1,animal2,animal3,animal4,animal5,animal6,animal7,animal8,animal9,animal10]
```

```python
#hacer la busqueda de la etiqueta h1 que corresponde al encabezado de cada animal
lista = []
for link in listado:
    request = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    soup = bs.BeautifulSoup(source, 'html.parser')
    encabezado = soup.find("body").find_all("h1")[0].text
    lista.append(encabezado)
i=1    
for h in lista:
    print(f'{i}. {h}')
    i+=1
```

```python
#textos que están en las etiquetas de negrilla.
for link in listado:
    request = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    soup = bs.BeautifulSoup(source, 'html.parser')
    negritas = soup.find("body").find_all("b")
    listadoNegritas =[]   
    for val in negritas:
        listadoNegritas.append(val.text)
   
    print(f'Página Web: {link}')
    print(f'Palabras en negrilla:')       
    i=1   
    for palabra in listadoNegritas:
        print(f'{i}. {palabra}')
        i+=1
```

```python
for link in listado:
    request = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    soup = bs.BeautifulSoup(source, 'html.parser')
    negritas = soup.find("body").find_all("b")
    listadoNegritas =[]   
    for val in negritas:
        listadoNegritas.append(val.text)
   
    print(f'Página Web: {link}')
    print(f'Palabras en negrilla:')       
    i=1   
    for palabra in listadoNegritas:
        resultado = re.sub(r"[áéíóú]", "*", palabra)
        print(f'{i}. {resultado}')
        i+=1
```
